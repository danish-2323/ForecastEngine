# ForecastEngine: Data Exploration

This notebook explores the time-series data to understand patterns, trends, and data quality.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette('husl')
```

## 1. Load and Inspect Data

```python
# Load sample data
data = pd.read_csv('../data/sample_data.csv')
data['date'] = pd.to_datetime(data['date'])

print(f"Dataset shape: {data.shape}")
print(f"Date range: {data['date'].min()} to {data['date'].max()}")
print(f"Columns: {list(data.columns)}")

data.head()
```

```python
# Data quality check
print("Missing values:")
print(data.isnull().sum())

print("\nData types:")
print(data.dtypes)

print("\nBasic statistics:")
data.describe()
```

## 2. Time Series Visualization

```python
# Plot main time series
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Main value over time
axes[0,0].plot(data['date'], data['value'], linewidth=2)
axes[0,0].set_title('Value Over Time')
axes[0,0].set_xlabel('Date')
axes[0,0].set_ylabel('Value')

# Price over time
axes[0,1].plot(data['date'], data['price'], color='orange', linewidth=2)
axes[0,1].set_title('Price Over Time')
axes[0,1].set_xlabel('Date')
axes[0,1].set_ylabel('Price')

# Temperature over time
axes[1,0].plot(data['date'], data['temperature'], color='red', linewidth=2)
axes[1,0].set_title('Temperature Over Time')
axes[1,0].set_xlabel('Date')
axes[1,0].set_ylabel('Temperature')

# Promotion indicator
promotion_dates = data[data['promotion'] == 1]['date']
promotion_values = data[data['promotion'] == 1]['value']
axes[1,1].plot(data['date'], data['value'], alpha=0.7, linewidth=1)
axes[1,1].scatter(promotion_dates, promotion_values, color='red', s=50, alpha=0.8)
axes[1,1].set_title('Value with Promotion Days (Red Dots)')
axes[1,1].set_xlabel('Date')
axes[1,1].set_ylabel('Value')

plt.tight_layout()
plt.show()
```

## 3. Seasonality Analysis

```python
# Add time-based features for analysis
data['day_of_week'] = data['date'].dt.dayofweek
data['month'] = data['date'].dt.month
data['day_of_month'] = data['date'].dt.day

# Weekly patterns
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Day of week pattern
weekly_avg = data.groupby('day_of_week')['value'].mean()
axes[0].bar(range(7), weekly_avg.values)
axes[0].set_title('Average Value by Day of Week')
axes[0].set_xlabel('Day of Week (0=Monday)')
axes[0].set_ylabel('Average Value')
axes[0].set_xticks(range(7))
axes[0].set_xticklabels(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])

# Monthly pattern
monthly_avg = data.groupby('month')['value'].mean()
axes[1].bar(monthly_avg.index, monthly_avg.values)
axes[1].set_title('Average Value by Month')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Average Value')

# Promotion impact
promo_impact = data.groupby('promotion')['value'].mean()
axes[2].bar(['No Promotion', 'Promotion'], promo_impact.values, color=['blue', 'red'])
axes[2].set_title('Average Value by Promotion Status')
axes[2].set_ylabel('Average Value')

plt.tight_layout()
plt.show()
```

## 4. Correlation Analysis

```python
# Correlation matrix
numeric_cols = ['value', 'price', 'promotion', 'temperature']
correlation_matrix = data[numeric_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=0.5)
plt.title('Feature Correlation Matrix')
plt.show()

print("Correlation with target variable 'value':")
correlations = data[numeric_cols].corr()['value'].sort_values(ascending=False)
print(correlations)
```

## 5. Trend and Stationarity

```python
# Decompose trend and seasonality (simplified)
from scipy import signal

# Calculate rolling statistics
window = 7
data['rolling_mean'] = data['value'].rolling(window=window).mean()
data['rolling_std'] = data['value'].rolling(window=window).std()

# Plot trend analysis
fig, axes = plt.subplots(2, 1, figsize=(15, 8))

# Original vs trend
axes[0].plot(data['date'], data['value'], label='Original', alpha=0.7)
axes[0].plot(data['date'], data['rolling_mean'], label=f'{window}-day Moving Average', linewidth=2)
axes[0].set_title('Value with Trend')
axes[0].set_ylabel('Value')
axes[0].legend()

# Rolling standard deviation
axes[1].plot(data['date'], data['rolling_std'], color='red', linewidth=2)
axes[1].set_title('Rolling Standard Deviation (Volatility)')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Standard Deviation')

plt.tight_layout()
plt.show()
```

## 6. Data Quality Summary

```python
# Summary statistics
print("=== DATA QUALITY SUMMARY ===")
print(f"Total records: {len(data)}")
print(f"Date range: {(data['date'].max() - data['date'].min()).days} days")
print(f"Missing values: {data.isnull().sum().sum()}")
print(f"Duplicate dates: {data['date'].duplicated().sum()}")

print("\n=== VALUE STATISTICS ===")
print(f"Mean: {data['value'].mean():.2f}")
print(f"Std: {data['value'].std():.2f}")
print(f"Min: {data['value'].min():.2f}")
print(f"Max: {data['value'].max():.2f}")
print(f"Growth rate: {((data['value'].iloc[-1] / data['value'].iloc[0]) - 1) * 100:.1f}%")

print("\n=== SEASONALITY INSIGHTS ===")
print(f"Strongest day-of-week effect: {weekly_avg.idxmax()} (value: {weekly_avg.max():.1f})")
print(f"Weakest day-of-week effect: {weekly_avg.idxmin()} (value: {weekly_avg.min():.1f})")
print(f"Promotion impact: +{(promo_impact[1] - promo_impact[0]):.1f} units")
print(f"Temperature correlation: {data['value'].corr(data['temperature']):.3f}")
```
