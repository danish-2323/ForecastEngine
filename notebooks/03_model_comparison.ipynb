# ForecastEngine: Model Comparison

This notebook compares different forecasting models and ensemble approaches.

```python
import sys
sys.path.append('../src')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from models.model_factory import ModelFactory, ARIMAModel, RandomForestModel, LinearModel
from ensemble.ensemble_manager import EnsembleManager
from evaluation.evaluator import ModelEvaluator
from feature_engineering.feature_builder import FeatureBuilder
from sklearn.metrics import mean_absolute_error, mean_squared_error
import warnings
warnings.filterwarnings('ignore')
```

## 1. Prepare Data and Features

```python
# Load and prepare data
data = pd.read_csv('../data/sample_data.csv')
data['date'] = pd.to_datetime(data['date'])

# Build features
config = {
    'lags': {'periods': [1, 2, 3, 7, 14]},
    'rolling_windows': {'windows': [7, 14], 'functions': ['mean', 'std']},
    'seasonality': {'enabled': True}
}

fb = FeatureBuilder(config)
features_data = fb.build_features(data, 'value', 'date', ['price', 'promotion', 'temperature'])

print(f"Features data shape: {features_data.shape}")
print(f"Date range: {features_data['date'].min()} to {features_data['date'].max()}")

# Split data
split_idx = int(len(features_data) * 0.8)
train_data = features_data.iloc[:split_idx]
test_data = features_data.iloc[split_idx:]

print(f"Train data: {len(train_data)} samples")
print(f"Test data: {len(test_data)} samples")
```

## 2. Individual Model Training and Evaluation

```python
# Initialize models
models = {
    'arima': ARIMAModel(),
    'random_forest': RandomForestModel({'random_forest': {'n_estimators': 100, 'max_depth': 10}}),
    'linear': LinearModel()
}

# Train models
trained_models = {}
for name, model in models.items():
    print(f"Training {name}...")
    trained_models[name] = model.fit(train_data, 'value')

print("All models trained successfully!")
```

```python
# Generate predictions
feature_columns = [col for col in test_data.columns 
                  if col not in ['value', 'date'] and not col.startswith('Unnamed')]

X_test = test_data[feature_columns]
y_test = test_data['value'].values

predictions = {}
for name, model in trained_models.items():
    try:
        pred = model.predict(X_test)
        predictions[name] = pred
        print(f"{name} predictions generated: {len(pred)} values")
    except Exception as e:
        print(f"Error with {name}: {e}")

print(f"\nActual values range: {y_test.min():.1f} to {y_test.max():.1f}")
for name, pred in predictions.items():
    print(f"{name} predictions range: {pred.min():.1f} to {pred.max():.1f}")
```

## 3. Model Performance Comparison

```python
# Calculate metrics for each model
def calculate_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    return {'MAE': mae, 'RMSE': rmse, 'MAPE': mape}

# Evaluate all models
results = {}
for name, pred in predictions.items():
    # Ensure same length
    min_len = min(len(y_test), len(pred))
    metrics = calculate_metrics(y_test[:min_len], pred[:min_len])
    results[name] = metrics

# Create results DataFrame
results_df = pd.DataFrame(results).T
results_df = results_df.round(3)

print("=== MODEL PERFORMANCE COMPARISON ===")
print(results_df)

# Find best model for each metric
print("\n=== BEST MODELS BY METRIC ===")
for metric in ['MAE', 'RMSE', 'MAPE']:
    best_model = results_df[metric].idxmin()
    best_value = results_df.loc[best_model, metric]
    print(f"Best {metric}: {best_model} ({best_value:.3f})")
```

```python
# Visualize performance comparison
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

metrics = ['MAE', 'RMSE', 'MAPE']
colors = ['blue', 'green', 'orange']

for i, metric in enumerate(metrics):
    values = results_df[metric]
    axes[i].bar(values.index, values.values, color=colors[i], alpha=0.7)
    axes[i].set_title(f'{metric} Comparison')
    axes[i].set_ylabel(metric)
    axes[i].tick_params(axis='x', rotation=45)
    
    # Highlight best performer
    best_idx = values.idxmin()
    best_pos = list(values.index).index(best_idx)
    axes[i].bar(best_idx, values[best_idx], color='red', alpha=0.8)

plt.tight_layout()
plt.show()
```

## 4. Ensemble Model Comparison

```python
# Test different ensemble methods
ensemble_configs = {
    'simple_average': {'method': 'simple_average'},
    'weighted_average': {'method': 'weighted_average'}
}

ensemble_results = {}

for ensemble_name, config in ensemble_configs.items():
    print(f"Testing {ensemble_name} ensemble...")
    
    # Create ensemble
    ensemble = EnsembleManager(config)
    ensemble.fit(trained_models, train_data, 'value')
    
    # Generate ensemble predictions
    ensemble_pred = ensemble.predict(X_test, horizon=len(X_test), models=trained_models)
    
    # Calculate metrics
    min_len = min(len(y_test), len(ensemble_pred))
    ensemble_metrics = calculate_metrics(y_test[:min_len], ensemble_pred[:min_len])
    ensemble_results[ensemble_name] = ensemble_metrics
    
    # Store predictions for visualization
    predictions[f'ensemble_{ensemble_name}'] = ensemble_pred
    
    print(f"Weights: {ensemble.get_model_weights()}")

# Add ensemble results to comparison
all_results = {**results, **ensemble_results}
all_results_df = pd.DataFrame(all_results).T.round(3)

print("\n=== COMPLETE PERFORMANCE COMPARISON ===")
print(all_results_df)

# Highlight best overall performer
best_overall = all_results_df['MAE'].idxmin()
print(f"\nBest overall model: {best_overall} (MAE: {all_results_df.loc[best_overall, 'MAE']:.3f})")
```

## 5. Prediction Visualization

```python
# Plot actual vs predicted for all models
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.ravel()

test_dates = test_data['date'].values
colors = ['blue', 'green', 'orange', 'red', 'purple']

# Plot individual models
model_names = list(predictions.keys())[:4]  # First 4 models

for i, model_name in enumerate(model_names):
    pred = predictions[model_name]
    min_len = min(len(y_test), len(pred), len(test_dates))
    
    axes[i].plot(test_dates[:min_len], y_test[:min_len], label='Actual', color='black', linewidth=2)
    axes[i].plot(test_dates[:min_len], pred[:min_len], label='Predicted', color=colors[i], linewidth=2, alpha=0.8)
    axes[i].set_title(f'{model_name.replace("_", " ").title()} Model')
    axes[i].set_xlabel('Date')
    axes[i].set_ylabel('Value')
    axes[i].legend()
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Compare all models on same plot
plt.figure(figsize=(15, 8))

# Plot actual values
plt.plot(test_dates, y_test, label='Actual', color='black', linewidth=3, alpha=0.8)

# Plot predictions from top 3 models
top_models = all_results_df.nsmallest(3, 'MAE').index.tolist()

for i, model_name in enumerate(top_models):
    if model_name in predictions:
        pred = predictions[model_name]
        min_len = min(len(y_test), len(pred), len(test_dates))
        plt.plot(test_dates[:min_len], pred[:min_len], 
                label=f'{model_name.replace("_", " ").title()}', 
                color=colors[i], linewidth=2, alpha=0.7)

plt.title('Top 3 Models vs Actual Values')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 6. Model Analysis Summary

```python
# Performance improvement analysis
baseline_mae = all_results_df.loc['linear', 'MAE']  # Use linear as baseline
all_results_df['MAE_Improvement'] = ((baseline_mae - all_results_df['MAE']) / baseline_mae * 100).round(1)

print("=== FINAL MODEL COMPARISON SUMMARY ===")
print(all_results_df.sort_values('MAE'))

print("\n=== KEY INSIGHTS ===")
best_model = all_results_df['MAE'].idxmin()
best_mae = all_results_df.loc[best_model, 'MAE']
best_improvement = all_results_df.loc[best_model, 'MAE_Improvement']

print(f"1. Best performing model: {best_model}")
print(f"2. Best MAE: {best_mae:.3f}")
print(f"3. Improvement over baseline: {best_improvement:.1f}%")

# Ensemble analysis
ensemble_models = [name for name in all_results_df.index if 'ensemble' in name]
if ensemble_models:
    best_ensemble = min(ensemble_models, key=lambda x: all_results_df.loc[x, 'MAE'])
    ensemble_mae = all_results_df.loc[best_ensemble, 'MAE']
    
    individual_models = [name for name in all_results_df.index if 'ensemble' not in name]
    best_individual = min(individual_models, key=lambda x: all_results_df.loc[x, 'MAE'])
    individual_mae = all_results_df.loc[best_individual, 'MAE']
    
    ensemble_benefit = ((individual_mae - ensemble_mae) / individual_mae * 100)
    
    print(f"4. Best ensemble: {best_ensemble} (MAE: {ensemble_mae:.3f})")
    print(f"5. Best individual: {best_individual} (MAE: {individual_mae:.3f})")
    print(f"6. Ensemble benefit: {ensemble_benefit:.1f}%")

print("\n=== RECOMMENDATIONS ===")
print(f"• Use {best_model} for production deployment")
print(f"• Expected forecast accuracy: ±{best_mae:.1f} units")
if ensemble_benefit > 0:
    print(f"• Ensemble provides {ensemble_benefit:.1f}% improvement over single models")
else:
    print(f"• Single model performs better than ensemble")
print(f"• Monitor model performance and retrain if MAE exceeds {best_mae * 1.2:.1f}")
```
